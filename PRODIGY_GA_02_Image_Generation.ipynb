```python
import torch
from diffusers import StableDiffusionPipeline
```

```python
model_id = "runwayml/stable-diffusion-v1-5"

pipe = StableDiffusionPipeline.from_pretrained(
    model_id,
    torch_dtype=torch.float16
)

pipe = pipe.to("cuda")
```

```python
prompt = "A breathtaking futuristic smart city at sunset, tall glass skyscrapers with glowing neon lights, numerous flying cars moving through the sky, elevated highways, holographic billboards, lush rooftop gardens, reflective buildings, cinematic lighting, ultra realistic, highly detailed, vibrant colors, 8k, masterpiece, digital art"

image = pipe(prompt).images[0]
image.save("image1.png")
image
```
